```python
import pandas as pd
from sklearn.ensemble import RandomForestRegressor
import numpy as np
import sklearn.metrics
```

```python
df = pd.read_csv("..//data_mart//reporting_cb.csv")
hold_out_year = 2017
```

```python
test_df = df.loc[df['draft_year'] == hold_out_year]
```

```python
bool_series = pd.notnull(df['max_madden'])
train_df = df[bool_series]
train_df = train_df.loc[train_df['draft_year'] != hold_out_year]
```

```python
cols = train_df.columns
for value in cols:
    print(value)
```

```python
numerical_columns = ['cb_hw_ratio_zscore', 'cb_forty_zscore', 'cb_vertical_zscore', 'cb_bench_zscore', 'cb_broad_zscore', 'cb_threecone_zscore','cb_shuttle_zscore'
                     ,'cb_football_spending_zscore', 'cb_ann_rain_inch_zscore', 
                    'cb_defense_ast_tackles_pg_cf_scaled_zscore', 'cb_defense_fum_forced_pg_cf_scaled_zscore','cb_defense_fum_rec_pg_cf_scaled_zscore','cb_defense_int_pg_cf_scaled_zscore','cb_defense_loss_tackles_pg_cf_scaled_zscore',
'cb_defense_pd_pg_cf_scaled_zscore','cb_defense_sacks_pg_cf_scaled_zscore','cb_defense_solo_tackes_pg_cf_scaled_zscore','cb_defense_tackles_pg_cf_scaled_zscore']
```

```python
X = train_df[numerical_columns]
y = train_df['max_madden']
```

```python
regr = RandomForestRegressor(random_state=0)
X = X.fillna(0)
regr.fit(X,y)
```

```python
test_df = test_df.fillna(0)
preds = regr.predict(test_df[numerical_columns])
```

```python
test_df['max_madden'] = test_df['max_madden'].replace(0, 65)
print(test_df[['full_name', 'max_madden']])
```

```python
sklearn.metrics.r2_score(list(test_df['max_madden']), preds)
```

```python
se = pd.Series(list(preds))
```

```python
test_df['preds'] = se.values
```

```python
result_df = test_df[['full_name', 'max_madden', 'preds', 'position_group_rank']]
result_df = result_df.rename(columns={'position_group_rank':'draft_position', 'preds':'predicted_madden'})
result_df.to_csv(r'C:\Users\B\Desktop\results.csv',index=False)
result_df
```
