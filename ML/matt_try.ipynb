```python
import pandas as pd
from sklearn.ensemble import RandomForestRegressor
import numpy as np
import sklearn.metrics
```

```python
df = pd.read_csv("..//data_mart//reporting_rb.csv")
hold_out_year = 2015
```

```python
print(df.columns)
```

```python
test_df = df.loc[df['draft_year'] == hold_out_year]
```

```python
bool_series = pd.notnull(df['max_madden'])
train_df = df[bool_series]
train_df = train_df.loc[train_df['draft_year'] != hold_out_year]
print(train_df)
```

```python
print(set(list(train_df['draft_year'])))
```

```python
numerical_columns = ['threecone_zscore',
       'shuttle_zscore', 'rushing_games_zscore', 'rushing_rec_td_zscore',
       'rushing_rec_yards_zscore','rushing_rec_td_pg_zscore', 'rushing_rec_yards_pg_zscore',
       'rushing_receptions_pg_zscore', 'rushing_rush_att_pg_zscore',
       'rushing_rush_td_pg_zscore', 'rushing_rush_yds_pg_zscore',
       'rushing_scrim_plays_pg_zscore', 'rushing_scrim_tds_pg_zscore',
       'rushing_scrim_yds_pg_zscore', 'football_spending_zscore',
       'coaches_salaries_zscore', 'ticket_sales_zscore']
```

```python
X = train_df[numerical_columns]
y = train_df['max_madden']
```

```python
regr = RandomForestRegressor(random_state=0)
X = X.fillna(0)
regr.fit(X,y)
```

```python
test_df = test_df.fillna(0)
preds = regr.predict(test_df[numerical_columns])
```

```python
test_df['max_madden'] = test_df['max_madden'].replace(0, 65)
print(test_df[['full_name', 'max_madden']])
```

```python
sklearn.metrics.r2_score(list(test_df['max_madden']), preds)
```

```python
se = pd.Series(list(preds))
```

```python
test_df['preds'] = se.values
```

```python
result_df = test_df[['full_name', 'max_madden', 'preds', 'position_group_rank']]
```
