```python
import pandas as pd
```

```python
input_file = "rb_rank_of_pics.csv"
output_file = input_file.replace(".csv", ".json")  #file for sankey chart

data=pd.read_csv(input_file)
data.drop(['Unnamed: 0'], axis=1, inplace=True)
data.sort_values(by=['position_group_rank'],inplace=True)
data.head(3)
```

```python
data['position_group_rank'] = data['position_group_rank'].astype(int)
data['ElasticNetCV Bins'] = data['ElasticNetCV Bins'].astype(int)
data.head(3)
```

```python
colDraft='position_group_rank'
first_col = data.pop(colDraft)
data.insert(0, colDraft, first_col)
```

```python
data.head()
```

```python
data=data[['position_group_rank','full_name','madden_cat','ElasticNetCV Bins']]
data.head()
```

```python
data = data.reset_index(drop=True)
data.head()
```

```python
data['NAME']=data['position_group_rank'].astype(str)+' '+data['full_name']
data['MAX_MADDEN']=data['madden_cat'].astype(str)+' Max Madden'
data['PRED_MADDEN']=data['ElasticNetCV Bins'].astype(str)+' Predicted Madden'
data.head()
```

```python
data=data[:24]  #24 seems to look best in sankey - good space distribution
data
```

```python
i=0
nameNodes=[]
maxMaddenNodes=[]
predMaddenNodes=[]
nodesSorted=[]
links=[]
strJson=""
for index, row in data.iterrows():
    # parse from left to right, then down. Print each new node
    name= row['NAME']
    maxMadden= row['MAX_MADDEN']
    predMadden= row['PRED_MADDEN']
    linkName=row['full_name']
    
    if name not in nameNodes:
        nameNodes.append(name)
        nodesSorted.append(name)
        i+=1
        if maxMadden not in maxMaddenNodes:
            maxMaddenNodes.append(maxMadden) 
            nodesSorted.append(maxMadden)
            i+=1
        if predMadden not in predMaddenNodes:
            predMaddenNodes.append(predMadden)
            nodesSorted.append(predMadden)
            i+=1
        
    #get Node id
    nameIdx=nodesSorted.index(name)
    maxMaddeenIdx=nodesSorted.index(maxMadden)
    predMaddeneIdx=nodesSorted.index(predMadden)
    links.append([nameIdx,maxMaddeenIdx,predMaddeneIdx,linkName])
    #{"source":18,"target":15,"value":1},
    #strJson+="{}:{},{}:{}{}{}".format("{\"source\"", nameIdx, "\"target\"","\"", node,"\"},\n")

print(links)
```

```python
def createLinks(links):
    strJson=""
    strJson+="{}".format("\"links\":[\n")
    i=0
    for link in links:
        #{"source":0,"target":1,"value":1},
        strJson+="{}:{},{}:{},{}:{},{}:{},{}:{}{}".format("{\"id\"", i,"\"id_next\"", i+1,"\"source\"", link[0], "\"target\"", link[1],"\"value\":1,\"label\"","\""+link[3],"\"},\n")
        i+=1
        strJson+="{}:{},{}:{},{}:{},{}:{},{}:{}{}".format("{\"id\"", i,"\"id_next\"", i-1,"\"source\"", link[1], "\"target\"", link[2],"\"value\":1,\"label\"","\""+link[3],"\"},\n")
        i+=1
    strJson=strJson[:-2]   
    strJson+="\n"
    strJson+="{}".format("]}\n")     
    return strJson
```

```python
def createNodes(nodesSorted):
    strJson=""
    ind=0
    strJson="{}".format("{\n")
    strJson+="{}".format("\"nodes\":[\n")
    for node in nodesSorted:
        strJson+="{}:{},{}:{}{}{}".format("{\"node\"", ind, "\"name\"","\"", node,"\"},\n")
        ind+=1
    strJson=strJson[:-2]   
    strJson+="\n"
    strJson+="{}".format("],\n") 
    return strJson
#string[:-1]    
    
```

```python
strJsonLinks=createLinks(links)
print(strJsonLinks)
```

```python
strJsonNodes=createNodes(nodesSorted)
print(strJsonNodes)
```

```python
strJsonForModel=strJsonNodes+strJsonLinks
print(strJsonForModel)
```

```python
out_file = open(output_file, "w")
n = out_file.write(strJsonForModel)
out_file.close()
```
